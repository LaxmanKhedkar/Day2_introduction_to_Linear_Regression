```python
import pandas as pd  # to read the data 
import numpy as np   # to perform the numerical operation 
import matplotlib.pyplot as plt  # use for plot the grap/Visualization 
```

```python
df = pd.read_csv("Happyness_Score.csv")  # read the file 
```

```python
df   # display the data 
```

```python
df = df.drop(["Unnamed: 0","Happiness Score"], axis = 1)  # drop the two columns 
```

```python
df
```

```python
plt.scatter(df["Income"], df["New Happiness Score"]) # using two columns plot the graph
plt.xlabel("Income") # x-axis of the graph 
plt.ylabel("Happiness Score") # y-axis of the graph
plt.show() # to display the graph
```

```python
x = df[["Income"]] # to display the income data 
```

```python
x
```

```python
y = df[["New Happiness Score"]]  # to display the New Happiness Score data 
```

```python
y
```

```python
from sklearn.model_selection import train_test_split  # import the library and select the model 
from sklearn.linear_model import LinearRegression  # importing the linear regression model 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)  # x,y data are train test in two parts 
```

```python
x_train # to display the training data 
```

```python
y_train # to display the training data 
```

```python
linear_regression_model = LinearRegression()
```

```python
linear_regression_model.fit(x_train,y_train)
```

```python
linear_regression_model.predict([[33000.999]])
```

```python
x_test
```

```python
linear_regression_model.predict(x_test)
```

```python
plt.scatter(df["Income"],df["New Happiness Score"])
plt.plot(x_train,linear_regression_model.predict(x_train), color = "red")
plt.xlabel("Income")
plt.ylabel("Happiness Score")
plt.show()
```
